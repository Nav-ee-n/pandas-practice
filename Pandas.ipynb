Pandas


```python
import pandas as pd
import numpy as np
```

```python
# Dataframe is structured data - rows and columns
# series is columns in dataframe - building blocks of dataframe - labelled 1d array - series have titles
#
# series uses more memory than array
```

```python
# create an array with 10 elements
a=np.arange(1,11)
a

b=pd.Series(a)
b

b.ndim
b.size
b.shape
```

```python
b.head()          #print first 5 elements
b.head(6)         #by default its 5 but value can be given
```

```python
b.tail()          #print last 5 elements of the series
b.tail(6)         #print last elements upto given value
```

```python
# add two series

a=pd.Series([1,2,3,4,5])
b=pd.Series([10,20,30,40,50])
c=pd.concat([a,b],ignore_index=True)
c
```

```python
a.add(b)
a.sub(b)
a.mul(b)
a.div(b)
```

```python
# create a series with named index
a=pd.Series([1,2,3,4],index=["a","b","c","d"])
a
```

```python
# series can be created using a dictionary
d={101:"apple",102:"orange",103:"mango"}

e=pd.Series(d)
e
```

```python
s=pd.Series(4,index=[1,2,3,4,5,6])
s
```

```python
# accessing a single element in series
s=pd.Series([1,2,3,4,5])
s[2]
```

```python
s[2:4]
```

```python
s[[2,4]]
```

DataFrame

```python
l=[1,2,3,4,5]
df=pd.DataFrame(l)

df

df.ndim
df.shape
df.size
```

```python
# creating dataframe from nested list

l=[["apple",10],["orange",20]]
df=pd.DataFrame(l,columns=['fruits','price'])
df
```

```python
# create a dataframe with id name location and add 4 data points
l=[[1,"Rahul","Kochi"],[2,"Jishnu","Trivandrum"],[3,"Suraj","Rak"],[4,"Abrar","Toronto"]]
df=pd.DataFrame(l,columns=["id","name","location"])
df
```

```python
# creating a dataframe from dictionary and list

d={"name":["Hiran","Techno","Metro"],"location":["rak","abu","dub"],"age":[23,45,24]}

df=pd.DataFrame(d,index=["a","b","c"])

df
```

```python
# creating dataframe from list of dictionaries

l=[{"fruits":"apple","price":20},{"fruits":"orange","price":30},{"fruits":"mango","price":45}]

df=pd.DataFrame(l)

df
```

```python
# create a dataframe with id,name,age,salary,profession

l={"id":[101,102,103,104,105],"name":["Steven","Gerrard","Sergio","Ramos","Pique"],"age":[45,34,56,34,21],"salary":[12900,24000,45300,45700,34000],
   "profession":["Data Scientist","DB Administrator","Software Tester","Team Leader","Manager"]}

df=pd.DataFrame(l)

df

l["gender"]=["m","f","m","m","f"]           #adding a new column

df=pd.DataFrame(l)


df
```
